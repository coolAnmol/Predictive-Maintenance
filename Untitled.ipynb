# Model Training

```python
def add_RUL_column(df):
    train_grouped_by_unit = train_data1.groupby(by='unit number') 
    max_time_cycles = train_grouped_by_unit['time, in cycles'].max() 
    merged = df.merge(max_time_cycles.to_frame(name='max_time_cycle'), left_on='unit number',right_index=True)
    merged["RUL"] = merged["max_time_cycle"] - merged['time, in cycles']
    merged = merged.drop("max_time_cycle", axis=1) 
    return merged
```

```python
train_data1 = add_RUL_column(train_data1)
```

```python
from sklearn.model_selection import train_test_split
X_train=train_data1.drop(columns=["unit number", "time, in cycles", "operational setting 1", "operational setting 2", "operational setting 3"])
X_train, X_test, y_train, y_test=train_test_split(X_train,X_train['RUL'], test_size=0.3, random_state=42)
```

```python
y_train
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

#Droping the target variable
X_train.drop(columns=['RUL'], inplace=True)
X_test.drop(columns=['RUL'], inplace=True)

#Scaling X_train and X_test
X_train_s=scaler.fit_transform(X_train)
X_test_s=scaler.fit_transform(X_test)
#Conserve only the last occurence of each unit to match the length of rul1
X_valid = test_data1.groupby('unit number').last().reset_index().drop(columns=["unit number", "time, in cycles", "operational setting 1", "operational setting 2", "operational setting 3"])
#scaling X_valid
X_valid_s=scaler.fit_transform(X_valid)
```

### 1.Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
linreg=LinearRegression()
linreg.fit(X_train_s,y_train)
y_pred=linreg.predict(X_test_s)
mae=mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)
print("Mean absolute error", mae)
print("R2 Score", score)
plt.scatter(y_test,y_pred)
```

### 2.Support Vector Regressor

```python
from sklearn.svm import SVR
svr=SVR()
svr.fit(X_train_s, y_train)
y_pred1=svr.predict(X_test_s)
mae1=mean_absolute_error(y_test,y_pred1)
score1=r2_score(y_test,y_pred1)
print("Mean absolute error", mae1)
print("R2 Score", score1)
plt.scatter(y_test,y_pred1)
```

### 3. Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor
rfr=RandomForestRegressor()
rfr.fit(X_train_s, y_train)
y_pred2=rfr.predict(X_test_s)
mae2=mean_absolute_error(y_test,y_pred2)
score2=r2_score(y_test,y_pred2)
print("Mean absolute error", mae2)
print("R2 Score", score2)
plt.scatter(y_test,y_pred2)
```

### 4. Ridge Regression

```python
from sklearn.linear_model import Ridge

ridge=Ridge()
ridge.fit(X_train_s, y_train)
y_pred3=ridge.predict(X_test_s)
mae3=mean_absolute_error(y_test,y_pred3)
score3=r2_score(y_test,y_pred3)
print("Mean absolute error", mae3)
print("R2 Score", score3)
plt.scatter(y_test,y_pred3)
```

### 5. Lasso Regression

### 6. XG-Bosst Regressor
